```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
from collections import OrderedDict
print("Imports successful")
```

## Task 1 – Create the custom network

```python
edges = [(1, 3), (1, 2), (2, 3), (2, 5), (3, 5), (3, 4), (4, 5)]
CGraph = nx.Graph()
CGraph.add_edges_from(edges)

print(f"Given graph — nodes: {CGraph.number_of_nodes()}, edges: {CGraph.number_of_edges()}")
print(f"Edges: {sorted(CGraph.edges())}")

nx.draw(CGraph, with_labels=True, node_color="#66b3ff", edge_color="#555", node_size=800)
```

## Task 2 – Closeness centrality of custom network

$$
 C_{\text{close}}(u) = \frac{n - 1}{\sum_{v \ne u} d(u,v)}
$$

$$
 C_{\text{bet}}(u) = \sum_{s \ne u \ne t} \frac{\sigma_{st}(u)}{\sigma_{st}}
$$


```python
custom_closeness = nx.closeness_centrality(CGraph)
for node, val in sorted(custom_closeness.items()):
    print(f"Node {node}: closeness {val:.4f}")
```

## Task 3 – Betweenness centrality of custom network

```python
custom_betweenness = nx.betweenness_centrality(CGraph)
for node, val in sorted(custom_betweenness.items()):
    print(f"Node {node}: betweenness {val:.4f}")
```

## Task 4 – Display the Karate Club network

```python
karate = nx.karate_club_graph()
print(f"Karate Club graph — nodes: {karate.number_of_nodes()}, edges: {karate.number_of_edges()}")

plt.figure(figsize=(7, 6))
pos_karate = nx.spring_layout(karate, seed=42)
node_sizes = [karate.degree(n) * 80 for n in karate.nodes()]
nx.draw_networkx(karate, pos=pos_karate, with_labels=True, node_size=node_sizes,
                 node_color="#ffcc66", edge_color="#888", linewidths=0.5, font_size=8)
plt.title("Karate Club network (spring layout)")
plt.axis("off")
plt.show()
```

## Task 5 – Closeness centrality of the Karate Club network

```python
karate_closeness = nx.closeness_centrality(karate)
for node, val in sorted(karate_closeness.items()):
    print(f"Node {node}: closeness {val:.4f}")
```

## Task 6 – Betweenness centrality of the Karate Club network

```python
karate_betweenness = nx.betweenness_centrality(karate)
for node, val in sorted(karate_betweenness.items()):
    print(f"Node {node}: betweenness {val:.4f}")
```

## Task 7 – Most important node(s) by closeness and betweenness

```python
def top_nodes(metric_dict):
    max_val = max(metric_dict.values())
    nodes = [n for n, v in metric_dict.items() if v == max_val]
    return nodes, max_val

top_close_nodes, top_close_val = top_nodes(karate_closeness)
top_between_nodes, top_between_val = top_nodes(karate_betweenness)

print("Closeness centrality — highest:")
print(f"Nodes {top_close_nodes} with score {top_close_val:.4f}")

print("\nBetweenness centrality — highest:")
print(f"Nodes {top_between_nodes} with score {top_between_val:.4f}")
```
